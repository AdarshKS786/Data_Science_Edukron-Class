```python
x={1,2}
y={3,4}
print(x.issubset(y))
print(y.issubset(x))
print(x.issuperset(y))
print(y.issuperset(x))
print(x.isdisjoint(y))
print(y.isdisjoint(x))
```

```python
x={1,2,3,4,5}
y={4,5,6,7,8}
x.difference_update(y)
print(x)
print(y)
```

```python
x={1,2,3,4,5}
y={4,5,6,7,8}
x.intersection_update(y)
print(x)
print(y)
```
