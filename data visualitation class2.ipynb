```python
g = sns.JointGrid(data=df, x="waiting", y="duration")
g.plot_joint(sns.scatterplot)
g.plot_marginals(sns.kdeplot)  # works for both numeric and categorical
```

```python
sns.pairplot(df[['waiting','duration']])
```

```python
df_numeric = df[["duration", "waiting"]].dropna()
df_numeric.head()
```

```python
g = sns.PairGrid(df_numeric)

g.map_diag(sns.histplot, fill=True)
g.map_offdiag(sns.scatterplot)
plt.show()
```

```python
g = sns.PairGrid(df_numeric)
g.map_upper(sns.kdeplot, fill=True)
g.map_lower(sns.regplot)
g.map_diag(sns.histplot, fill=True)
plt.show()
```

```python
import pandas as pd
import plotly.express as px
import plotly.io as pio

# Force browser rendering
pio.renderers.default = "browser"

# Define your data
usa_data = pd.DataFrame({
    'state': ['CA', 'TX', 'NY', 'FL', 'IL'],
    'sales': [300, 250, 220, 210, 180]
})

# Create choropleth map
fig = px.choropleth(
    usa_data,
    locations='state',
    locationmode='USA-states',
    color='sales',
    scope='usa',
    title='Dummy Sales by US State',
    labels={'sales': 'Sales Amount'}
)

fig.show()
```
