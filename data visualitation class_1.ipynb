```python
x = np.arange(100)
fig, axs = plt.subplots(2, 2)

axs[0,0].plot(x, np.sin(x))
axs[0,0].set_title('Sine')

axs[0,1].plot(x, np.cos(x))
axs[0,1].set_title('Cosine')

axs [1,0].plot(x, np.random.random(100))
axs[1,0].set_title('Random')

axs [1,1].plot(x, np.log(x))
axs[1,1].set_title('Random')
plt.tight_layout()

plt.suptitle('Four plots', fontsize=20)
plt.show

```

```python
fig.savefig("four plots.png", dpi=1000)
```

```python
x = plt.axes(projection="3d")

a = np.arange(0, 50,0.01)
b = np.arange(0, 50, 0.01)
c = np.cos(a+b)

x.scatter( a, b, c )

plt.title("3D Scatter Plot", font="times new roman", fontsize=20)
plt.show()
```

```python
a = plt.axes(projection="3d")

x= np.arange( -5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

X, Y = np.meshgrid(x, y)

Z = np.sin(X) * np.cos(Y)

a.plot_surface(X, Y,Z)

plt.title("3D Scatter Plot", font="times new roman", fontsize=20)
plt.show()
```

```python
a = plt.axes(projection="3d")

x= np.arange( -5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

X, Y = np.meshgrid(x, y)

Z = np.sin(X) * np.cos(Y)

a.plot_surface(X, Y,Z, cmap='Spectral')

plt.title("3D Scatter Plot", font="times new roman", fontsize=20)
plt.show()
```
